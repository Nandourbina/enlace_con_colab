```python
import numpy as np
from matplotlib import pyplot as plt

#Generar datos aleatorios
ys = 200 + np.random.randn(100)
x = [x for x in range(len(ys))]

#Trazar el grÃ¡fico de lÃ­nea con color azul ('b')
plt.plot(x, ys, '-', color='blue')

#Rellenar el Ã¡rea debajo de la lÃ­nea con color Amarillo  ('yellow')
plt.fill_between(x, ys, 195, where=(ys > 195), facecolor='yellow', alpha=0.6)

#Agregar un tÃ­tulo al grÃ¡fico
plt.title=("Ejemplo de Rellenos y Alfa")

#Mostrar el grÃ¡fico
plt.show()
```
